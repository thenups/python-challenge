```python
# Modules
import os, csv, datetime
from operator import itemgetter
```

```python
# Set path for file
bankDataCvsPath1 = os.path.join("raw_data","budget_data_1.csv")
bankDataCvsPath2 = os.path.join("raw_data","budget_data_2.csv")
```

```python
# Declare list variables
month = []
revenue = []
```

```python
# Declare output varibales
nMonths = 0 #total number of months included in the dataset
totalRevenue = 0 #total amount of revenue gained over the entire period
revenueDelt = 0 #average change in revenue between months over the entire period
revInc = 0 #greatest increase in revenue (date and amount) over the entire period
revDec = 0 #greatest decrease in revenue (date and amount) over the entire period
```

```python
# Function to figure out is value is an actual number
def isNumber(n):
    try:
        int(n)
        return True
    except ValueError:
        return  False
```

```python
def appendValueList(col1,col2):
    month.append(col1)
    revenue.append(int(col2[1]))
```

```python
def combineFiles (dataFile):
    with open(dataFile, 'r', newline="") as csvfile:
        csvreader = csv.reader(csvfile, delimiter=',')

        dateFormats = ["%b-%y","%b-%Y"]

        for row in csvreader:

            if isNumber(row[1]):

                for formats in dateFormats:
                    try:
                        dateObject = datetime.datetime.strptime(row[0],formats)
                    except ValueError:
                        pass

                appendValueList(dateObject,row)
                
```

```python
combineFiles(bankDataCvsPath1)
combineFiles(bankDataCvsPath2)
```
